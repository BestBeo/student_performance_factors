```python
# Xác định các biến phân loại
categorical_vars = df.select_dtypes(include=['object']).columns

# Thống kê số lượng giá trị duy nhất của từng biến phân loại
unique_values_count = {}
for var in categorical_vars:
    unique_count = df[var].nunique()
    unique_values_count[var] = unique_count

# In kết quả
print("Số lượng giá trị duy nhất của từng biến phân loại:")
for var, count in unique_values_count.items():
    print(f"- {var}: {count} giá trị duy nhất")
```

```python
# Thống kê số lượng giá trị duy nhất của từng biến phân loại
unique_counts = df.select_dtypes(include=['object']).nunique()

# Lọc các biến phân loại có số lượng giá trị duy nhất nhỏ hơn 10
filtered_variables = unique_counts[unique_counts <= 10].index

# In kết quả
print("Các biến phân loại có số lượng giá trị duy nhất nhỏ hơn 10:")
print(filtered_variables)

# Thống kê số lượng (tần suất) và phần trăm của mỗi loại cho các biến phân loại nhỏ hơn hoặc bằng 10
for var in filtered_variables:
    value_counts = df[var].value_counts()
    percentage = df[var].value_counts(normalize=True) * 100

    # In kết quả
    print(f"\nBiến phân loại: {var}")
    print("Số lượng (Tần suất):")
    print(value_counts)
    print("\nPhần trăm:")
    print(percentage)
```

```python
# Tính toán ma trận tương quan
correlation_matrix = numeric_df.corr()

# Vẽ biểu đồ heatmap
plt.figure(figsize=(30, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Ma trận Tương quan')
plt.show()
```

```python
# Biểu đồ đếm số lượng rủi ro giao hàng trễ
plt.figure(figsize=(8,6))
sns.countplot(x='Late_delivery_risk', data=df)
plt.title('Phân phối Rủi ro Giao hàng Trễ')
plt.show()

# Rủi ro giao hàng trễ theo các phương thức giao hàng khác nhau
plt.figure(figsize=(10,6))
sns.countplot(x='Shipping Mode', hue='Late_delivery_risk', data=df)
plt.title('Rủi ro Giao hàng Trễ theo Phương thức Giao hàng')
plt.show()
```

```python
# Chuyển đổi ngày đơn hàng sang dạng datetime
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])

# Vẽ biểu đồ doanh thu theo thời gian
df.groupby(df['order date (DateOrders)'].dt.to_period("M"))['Sales'].sum().plot(kind='line', figsize=(12,6))
plt.title('Xu hướng Doanh thu Hàng tháng')
plt.xlabel('Tháng')
plt.ylabel('Doanh thu')
plt.show()
```

```python
# Tính tổng doanh thu theo từng phương thức giao hàng
total_sales_by_shipping_mode = df.groupby('Shipping Mode')['Sales'].sum()
print(total_sales_by_shipping_mode)

```

```python
# Vẽ biểu đồ cột cho tổng doanh thu theo phương thức giao hàng
plt.figure(figsize=(10, 6))
total_sales_by_shipping_mode.plot(kind='bar')
plt.title('Doanh thu theo Phương thức Giao hàng')
plt.xlabel('Phương thức Giao hàng')
plt.ylabel('Doanh thu')
plt.xticks(rotation=45)  # Xoay nhãn trục x để dễ đọc
plt.show()
```
